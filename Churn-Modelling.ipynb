```python
data= pd.read_csv('Churn-Modelling.csv',index_col=0)
data.head()
```

```python
data.shape
```

```python
target= data['Exited']
features = data
features = features.drop('Exited', axis = 1)
features = features.drop('Surname', axis = 1)
features = features.drop('CustomerId', axis = 1)
features.shape
features.head()
```

```python
features['Gender']=features['Gender'].apply(lambda x: 1 if x == "Male" else 0)
```

```python
features.head(10)
```

```python

features_final = pd.get_dummies(features['Geography'])
# Print the number of features after one-hot encoding
encoded = list(features_final.columns)
print("{} total features after one-hot encoding.".format(len(encoded)))
# Uncomment the following line to see the encoded feature names
print(encoded)
```

```python
features_final.head(10)
```

```python
features['France']= features_final['France']
features['Germany']= features_final['Germany']
features['Spain']= features_final['Spain']
features.head(10)
```

```python
features.head(10)
```

```python
features = features.drop('Geography', axis = 1)

```

```python
features.head(30)
```

```python
X=features
y=target
print(X.shape)
print(y.shape)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y, size=.25 )
```

```python
y.head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model = SVC(kernel='rbf', gamma=27)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)
#cross_val_score(model(alpha=float(alpha), random_state=2),
 #                          X_train, y_train, scoring='mae', cv=5).mean()
scores = cross_val_score(model,X,y,cv=10,scoring='accuracy' )
print (scores.mean())
```
